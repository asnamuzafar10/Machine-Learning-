```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("ECG200_TRAIN.csv",delimiter="  ")
df
```

```python
#seperating the dataset on the basis of positive and negative labels
positive = df[df['-1.0000000e+00'] == 1.0]
negative = df[df['-1.0000000e+00'] == -1.0]
```

```python
split1 = 69//4
split2 = 30//4
print("\nSplitting Factor 1: ", split1, "\nSplitting Factor 2: ", split2)
```

```python
#concatenating the dataframes following their splitting factor
df1 = pd.concat([positive[:split1], negative[:split2]])
df2 = pd.concat([positive[split1:2*split1], negative[split2:2*split2]])
df3 = pd.concat([positive[2*split1:3*split1], negative[2*split2:3*split2]])
df4 = pd.concat([positive[3*split1:70], negative[3*split2:31]])
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils import shuffle
from sklearn.model_selection import KFold
df1 = shuffle(df1)
df2 = shuffle(df2)
df3 = shuffle(df3)
df4 = shuffle(df4)
```

```python
x = pd.concat([df1, df2, df3])
x
```

```python
ac1 = []
ac2 = []
```

```python
x = pd.concat([df1, df2, df3], axis=0)
y = df4
x_train = x.drop('-1.0000000e+00',axis=1)
y_train = x["-1.0000000e+00"]
x_test = y.drop('-1.0000000e+00',axis=1)
y_test = y["-1.0000000e+00"]

c2 = tree.DecisionTreeClassifier()
c2 = c2.fit(x_train, y_train)
p = c2.predict(x_test)
c2.score(x_test,y_test)

print ("Accuracy : " , accuracy_score(y_test,p)*100)  
ac1.append(accuracy_score(y_test,p)*100)
print("Report : \n", classification_report(y_test, p))
print("F1 Score : ",f1_score(y_test, p, average='macro')*100)
```

```python
#using sklearn for the same process
x = df.drop('-1.0000000e+00',axis=1)
y = df['-1.0000000e+00']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 39)
c2 = tree.DecisionTreeClassifier()
c2 = c2.fit(x_train, y_train)
p = c2.predict(x_test)
c2.score(x_test,y_test)

print ("Accuracy : " , accuracy_score(y_test,p)*100)
print("Report : \n", classification_report(y_test, p))
print("F1 Score : ",f1_score(y_test, p, average='macro')*100)
```

```python
#Applying kfold cross validation
kf = KFold(n_splits = 4,shuffle = True)
for i, j in kf.split(x): 
    x_train = x.iloc[i]
    x_test  = x.iloc[j]
    y_train = y.iloc[i]
    y_test  = y.iloc[j]    
    c2 = tree.DecisionTreeClassifier()
    c2 = c2.fit(x_train, y_train)
    p = c2.predict(x_test)
    c2.score(x_test,y_test)

    print ("Accuracy : " , accuracy_score(y_test,p)*100)
    ac2.append(accuracy_score(y_test,p)*100)
```
